## Analise Exploratoria dos Dados

## Grafico de Pareto

```python
#Criação da tabela de referencia para o grafico
prt = pd.DataFrame(data5['ID_Ocorrencia'].value_counts()).reset_index()
prt.rename(columns={'index':'ID_Ocorrencia','ID_Ocorrencia':'Frequencia'},inplace = True)
prt
```

```python
#Calculo das colunas 'Percentual' e 'Percentual Acumulado'
prt['Percentual'] = (prt['Frequencia']/prt['Frequencia'].sum())*100
prt['PercentualAcumulado'] = prt['Percentual'].cumsum()
prt
```

```python
#Adicionar legenda na tabela
prt = prt.merge(data1,on='ID_Ocorrencia',how='left')
prt
```

```python
#Grafico de Pareto
fig,ax = plt.subplots(figsize=(20,10))
ax.bar(prt['Ocorrencia'],prt['Frequencia'],color='Blue')
ax.set_title("Pareto Ocorrencias",fontsize=40)
ax2 = ax.twinx()
ax2.plot(prt['Ocorrencia'],prt['PercentualAcumulado'],color = 'red',marker='D',ms=7,label='Pareto')
ax2.yaxis.set_major_formatter(PercentFormatter())
ax.tick_params(axis="y", colors="black")
ax2.tick_params(axis="y", colors="black")
ax2.set_ylim([0,110])
for tick in ax.get_xticklabels():
    tick.set_rotation(90)
gap = 2.5
for i in range(len(prt['PercentualAcumulado'])):
    ax2.annotate("%.2f" % prt['PercentualAcumulado'][i], (prt["Ocorrencia"][i], prt['PercentualAcumulado'][i] + gap))
plt.legend()
plt.show()
```

## OEE

```python
#Criação de um novo dataset usando o df.merge() para ser usado como base para o calculo do OEE
data6 = data5.merge(data4[['ID_OP','ID_Produto']], on="ID_OP",how='left').merge(data3[['ID_Produto','Pecas_Hora']],on='ID_Produto',how='left')
data6['ProdIdeal'] = data6['Pecas_Hora']*data6['DifHoras']
data6
```

```python
#disponibilidade
TotalHoras = data6['DifHoras'].sum()
HorasParadas = data6[data6['ID_Ocorrencia'].notnull()]['DifHoras'].sum()
HorasProduzidas = data6[data6['ID_Ocorrencia'].isnull()]['DifHoras'].sum()
disp = (HorasProduzidas / TotalHoras)
disp
print(f"%Disponibilidade: {disp:.2%}")
```

```python
#Verificação de Variaveis
print(TotalHoras)
print(HorasParadas)
print(HorasProduzidas)
```

```python
#Qualidade
QtdBom = data6['QtdeProduzida'].sum()
QtdRuim = data6['QtdeRefugada'].sum()
QtdProd = QtdBom + QtdRuim
quali = QtdBom / QtdProd
quali
print(f"%Qualidade: {quali:.2%}")
```

```python
#Verificação de Variaveis
print(QtdBom)
print(QtdRuim)
print(QtdProd)
```

```python
#produtividade
#Calculo quantidade teorica
QtdTeorica = data6.loc[data6['ID_Ocorrencia'].isnull()]['ProdIdeal'].sum()
#calculo do indicador de produtividade
prod = QtdBom / QtdTeorica
print(f"%Produtividade: {prod:.2%}")
```

```python
#Verificação de Variaveis
print(QtdBom)
print(QtdTeorica)
```

```python
#OEE

OEE = prod*quali*disp
print(f"OEE: {OEE:.2%}")
```

## Analise do OEE

```python
print(f"%Produtividade: {prod:.2%}")
print(f"%Qualidade: {quali:.2%}")
print(f"%Disponibilidade: {disp:.2%}")
```

```python
#Função para avaliar se o OEE é de classe mundial

def classemundial(OEE,prod,quali,disp):
    if OEE >= 0.85:
        print('OEE de classe mundial')
    else:
        print('OEE Não é de classe mundial')
    if prod >= 0.95:
        print('Produtividade de classe mundial')
    else:
        print('Produtividade não é de classe mundial')
    if quali >= 0.99:
        print('Qualidade de classe mundial')
    else:
        print('Qualidade não é de classe mundial')
    if disp >= 0.90:
        print('Disponibilidade de classe mundial')
    else:
        print('Disponibilidade não é de classe mundial')
```

```python
classemundial(OEE,prod,quali,disp)
```
